```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

## 1.General statistics Plot(Pair plot) 

```python
sns.pairplot(iris,hue='species')
plt.show()
```

## 2.Pie Plot for Species Frequency

```python
species_counts=iris['species'].value_counts()
labels=iris['species'].unique()
plt.figure(figsize=(7,7))
explode=[0.01,0.01,0.01]
plt.pie(species_counts,explode=explode,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('Species Frequency in Iris Dataset')
plt.tight_layout()
plt.legend(labels=labels)
plt.show()
```

## 3.Scatter plot

To find the relationship between 2 columns we use sactter plot

#### Relationship between Sepal Length and Sepal Width

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal length in cm')
plt.ylabel('Sepal width in cm')
plt.show()
```

## 4.Distribution of Sepal and Petal Features

```python
sns.pairplot(iris,hue='species')
plt.show()
```

## 5.Joint plot

```python
sns.jointplot(x='sepal_length',y='sepal_width',hue='species',data=iris,kind='scatter')
plt.show()
```

## 6.KDE Plot for Setosa species

```python
setosa=iris[iris['species']=='setosa']
sns.kdeplot(x='sepal_length',y='sepal_width',data=setosa,fill=True)
plt.title('KDE Plot for Setosa species(sepal length vs sepal width)')
```

```python
sns.kdeplot(x='petal_length',y='petal_width',data=setosa,fill=True)
plt.title('KDE Plot for Setosa species(petal_length vs petal width)')
```
